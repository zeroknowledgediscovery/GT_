```python
import pandas as pd
import numpy as np
from scipy import stats

#using student t to calculate confidence of mean
def getT(M,alpha):
    return stats.t.ppf(1-0.5*(1-alpha), 4)

def getDF(X,alpha=.9,SAVE=None):
    M=X.index.size
    X_=pd.DataFrame(X.mean(),columns=['mn'])
    eps=(X.std()/np.sqrt(M))*getT(M,alpha)
    X_['U']=X_.mn+eps
    X_['L']=X_.mn-eps
    if SAVE is not None:
        X_.to_csv(SAVE)
    return X_
```
